```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel("This is my x Axis")
plt.ylabel("This is my y-Axis")
plt.title("This is my x vs y graph")
```

```python
plt.scatter(x,y,c='#223BDC',alpha=.9)
plt.xlabel("This is my x Axis")
plt.ylabel("This is my y-Axis")
plt.title("This is my x vs y graph")
```

```python
plt.scatter(x,y,c='#223BDC',alpha=.9,marker='X')
plt.xlabel("This is my x Axis")
plt.ylabel("This is my y-Axis")
plt.title("This is my x vs y graph")
plt.grid()
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y,c='#223BDC',alpha=.9,marker='X')
plt.xlabel("This is my x Axis")
plt.ylabel("This is my y-Axis")
plt.title("This is my x vs y graph")
plt.grid()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection ='3d')
ax.scatter3D(x,y,z,c='r',alpha=0.8)
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.show()
```
